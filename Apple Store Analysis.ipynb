```python
games_music = []

for q in applestore[1:]:
    prime = q[11]
    rating = float(q[7])
    price = float(q[4])
    if price == 0 and (prime == "Games" or prime == "Music"):
        games_music.append(rating)
```

```python
s6 = sum(games_music)
s6
```

```python
paid_genre = {}

for x in applestore[1:]:
    prime = x[11]
    price = float(x[4])
    if price != 0:
        if prime in paid_genre:
            paid_genre[prime] += 1
        else:
            paid_genre[prime] = 1
```
